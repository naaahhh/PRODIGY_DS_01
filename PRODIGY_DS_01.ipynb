```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns

df = pd.read_csv("population.csv", skiprows = 4)
```

```python
df.head()
```

```python
df = df.iloc[:, :-1]
df.head()
```

```python
#selecting a specific year and filtering the data for the selected year
df_2023 = df[['Country Name', '2023']].dropna()
```

```python
df_2023.columns = ['Country', 'Population']
```

```python
df_2023.head()
```

```python
df_2023['Population'] = pd.to_numeric(df_2023['Population'], errors='coerce')
df_2023 = df_2023.sort_values(by='Population')
sorted_df_2023 = df_2023.head(10)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Population', y='Country', data=sorted_df_2023, palette='viridis')
```
